# Berkeley PD Stop Data Analysis

A look at who Berkeley PD stops where

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("berkeleypd.csv")
df1 = df.iloc[:,0:5]
df_2019 = df1[df['Call Date/Time'].str.contains('2019', na=False)]
df_2019.dropna()
```

```python
df_2019_black = df_2019[df['Dispositions'].str.startswith('B', na=False)]
df_2019_white = df_2019[df['Dispositions'].str.startswith('W', na=False)]
```

```python
df_2019_men = df_2019.loc[df_2019['Dispositions'].str[1] == 'M'] 
df_2019_men_black = df_2019_men[df['Dispositions'].str.startswith('B', na=False)]
df_2019_men_white = df_2019_men[df['Dispositions'].str.startswith('W', na=False)]
```

```python
#arrests
df_2019_arrests = df_2019.loc[df_2019['Dispositions'].str[4] == 'A'] 
df_2019_arrests_black = df_2019_arrests[df['Dispositions'].str.startswith('B', na=False)]
df_2019_arrests_white = df_2019_arrests[df['Dispositions'].str.startswith('W', na=False)]

df_2019_arrests_men = df_2019_arrests.loc[df_2019['Dispositions'].str[1] == 'M'] 
df_2019_arrests_men_black = df_2019_arrests_men[df['Dispositions'].str.startswith('B', na=False)]
df_2019_arrests_men_white = df_2019_arrests_men[df['Dispositions'].str.startswith('W', na=False)]
```

```python
#types


#investigations
df_2019_investigations = df_2019.loc[df_2019['Dispositions'].str[3] == 'I'] 
df_2019_investigations_black = df_2019_investigations[df['Dispositions'].str.startswith('B', na=False)]
df_2019_investigations_white = df_2019_investigations[df['Dispositions'].str.startswith('W', na=False)]


#Traffic
df_2019_traffic = df_2019.loc[df_2019['Dispositions'].str[3] == 'T'] 
df_2019_traffic_black = df_2019_traffic[df['Dispositions'].str.startswith('B', na=False)]
df_2019_traffic_white = df_2019_traffic[df['Dispositions'].str.startswith('W', na=False)]

#Reasonable Suspicion
df_2019_reasonable = df_2019.loc[df_2019['Dispositions'].str[3] == 'R'] 
df_2019_reasonable_black = df_2019_reasonable[df['Dispositions'].str.startswith('B', na=False)]
df_2019_reasonable_white = df_2019_reasonable[df['Dispositions'].str.startswith('W', na=False)]

#Probation
df_2019_probation = df_2019.loc[df_2019['Dispositions'].str[3] == 'K'] 
df_2019_probation_black = df_2019_probation[df['Dispositions'].str.startswith('B', na=False)]
df_2019_probation_white = df_2019_probation[df['Dispositions'].str.startswith('W', na=False)]

#Wanted
df_2019_wanted = df_2019.loc[df_2019['Dispositions'].str[3] == 'W'] 
df_2019_wanted_black = df_2019_wanted[df['Dispositions'].str.startswith('B', na=False)]
df_2019_wanted_white = df_2019_wanted[df['Dispositions'].str.startswith('W', na=False)]

#Car Searched
df_2019_carsearched = df_2019.loc[df_2019['Dispositions'].str[5] == 'S'] 
df_2019_carsearched_black = df_2019_carsearched[df['Dispositions'].str.startswith('B', na=False)]
df_2019_carsearched_white = df_2019_carsearched[df['Dispositions'].str.startswith('W', na=False)]


#df_2019_carsearched.head()
len(df_2019_carsearched_white)
```
